```python
ventas = [
    [12000, 15000, 10000],
    [8000, 9000, 7000],
    [20000, 18000, 15000]
]

print("Ventas registradas:")

for i in range(3):
    print(f"Vendedor {i+1}: {ventas[i]}")

print("\nTotal por vendedor:")

mayor_total = 0
mejor_vendedor = 0

for i in range(3):
    total = sum(ventas[i])

    print(f"Vendedor {i+1}: ${total}")

    if total < 30000:
        print("Alerta: ventas menores a $30.000")

    if total > mayor_total:
        mayor_total = total
        mejor_vendedor = i + 1

print(f"\nEl vendedor con más ventas fue el Vendedor {mejor_vendedor}")
print(f"Total vendido: ${mayor_total}")
```
