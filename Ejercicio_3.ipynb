```python
edad = int(input("Ingrese su edad"))
tarjeta = input("¿Tienes tarjeta de soscio? (si/no)")
monto = int(input("Ingrese el monto de su compra"))
descuento = 15

if monto > 10000 and (edad >= 60 or tarjeta == "si"):
    print("Obtiene descuento", "\nSu monto actual con descuento integrado es:", monto - (monto * (descuento /100)))
else:
    print("No obtiene descuento", "\nSu monto actual es:", monto)
```
